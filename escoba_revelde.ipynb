```python
viajes = [
{"escoba":"Nimbus","km":[12,18,20]},
{"escoba":"Cometa","km":[8,7,10]},
{"escoba":"Rayo","km":[30,25,28]}
]
distancias = {
v["escoba"]:max(v["km"])
for v in viajes
}
print(distancias)
```

```python
viajes = [
{"escoba":"Nimbus","km":[12,18,20]},
{"escoba":"Cometa","km":[8,7,10]},
{"escoba":"Rayo","km":[30,25,28]}
]
distancias = {
v["escoba"]:sum(v["km"])
for v in viajes
}
print(distancias)
```
